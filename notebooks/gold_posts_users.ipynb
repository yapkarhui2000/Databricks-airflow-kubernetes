```python
stg_posts_df = spark.table("default.stg_posts")
raw_users_df = spark.table("default.raw_users")
```

```python
import pyspark.sql.functions as F
from pyspark.sql import DataFrame
def post_users_OBT(stg_posts_df:DataFrame, raw_users_df:DataFrame):
    return (
        stg_posts_df
        .alias("posts")
        .withColumnRenamed("CreationDate", "PostCreationDate")
        .join(
            raw_users_df.withColumnRenamed("CreationDate", "UserCreationDate").alias("users"),
            on=F.col("OwnerUserId") == F.col("users.Id"),
            how="left" 
        )
    )
marts_post_user = post_users_OBT(stg_posts_df, raw_users_df)
marts_post_user.write.mode("overwrite").saveAsTable("marts_post_user")
```
