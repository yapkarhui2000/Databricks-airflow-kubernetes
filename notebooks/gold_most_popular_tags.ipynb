```python
stg_posts_df = spark.table("default.stg_posts")
```

```python
 import pyspark.sql.functions as F
 from pyspark.sql import DataFrame

def most_tag_in_posts(stg_posts_df:DataFrame):
    return (
        stg_posts_df
        .withColumn("tagexploded",F.explode("TagsArray"))
        .groupBy("tagexploded")
        .agg(F.count("PostId").alias("count"))
        .orderBy(F.desc("count"))
    )
marts_top_tags_df = most_tag_in_posts(stg_posts_df)
marts_top_tags_df.write.mode("overwrite").saveAsTable("marts_top_tags")
    
```
